```python
years = []
for year in range(1819,0,-101):
    years.append(str(year))
years.append("9900")
for year in range(9899, 7675, -101):
    years.append(str(year))

# print(len(years))
# print(years)
```

```python
def getFileName(year):
    if int(year) > 1000:
        return year + "_misc.csv"
    elif int(year) > 100:
        return "0" + year + "_misc.csv"
    elif year == "1":
        return "000" + year + "_misc.csv"
```

```python
def getTeamYear(year):
    if int(year) >= 7677:
        return "19" + year[:2] + "/" + year[2:] + " "
    elif int(year) >= 1011:
        return "20" + year[:2] + "/" + year[2:] + " "
    elif int(year) >= 102:
        return "200" + year[:1] + "/" + year[1:] + " "
    elif year == "1":
        return "2000/01"
```

```python
import csv

data_CLEANED = []

for year in years:
    # open file
    with open(getFileName(year)) as f:
        data = []
    
        for row in csv.DictReader(f):
            data.append(dict(row))
    # clean data        
    for row in data:
        row["ORtg"] = float(row["ORtg"])
        row["DRtg"] = float(row["DRtg"])
        if (row["Team"] != "League Average"):
            row["W"] = int(row["W"])
            row["L"] = int(row["L"])
    # calculate and add new fields
    la_row = len(data) - 1
    average_ORtg = data[la_row]["ORtg"]
    average_DRtg = data[la_row]["DRtg"]
    for row in data[:la_row]:
        new_row = {}
        team_year = getTeamYear(year)
        if "*" in row["Team"]:
            new_row["Team"] =  team_year + row["Team"][:len(row["Team"]) - 1]
            new_row["Playoffs"] = True
        else:
            new_row["Team"] = team_year + row["Team"]
            new_row["Playoffs"] = False
        new_row["ORtg+"] = round(row["ORtg"] / average_ORtg * 100, 1)
        new_row["DRtg+"] = round(average_DRtg / row["DRtg"] * 100, 1)
    
        wins = row["W"]
        losses = row["L"]
        games_played = wins + losses
    
        new_row["PRW"] = round(wins / games_played * 82, 1)
        new_row["PRL"] = round(losses / games_played * 82, 1)
        data_CLEANED.append(new_row)

print(len(data_CLEANED))
display(data_CLEANED)
```

```python
csv_columns = ["Team", "Playoffs", "ORtg+", "DRtg+", "PRW", "PRL"]

csv_file = "data_rs_CLEANED.csv"
try:
    with open(csv_file, 'w') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
        writer.writeheader()
        for row in data_CLEANED:
            writer.writerow(row)
except IOError:
    print("I/O error")
```
