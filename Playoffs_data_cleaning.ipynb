```python
import csv

data_playoffs_CLEANED = []
data_champions_CLEANED = []
for year in range(2019, 1976, -1):
    with open(str(year) + "_playoffs.csv") as f:
        data = []
        for row in csv.DictReader(f):
            data.append(dict(row))
#     display(data)
            
    total_ortg = 0
    total_drtg = 0
    total_teams = 0
    for row in data:
        row["ORtg"] = float(row["ORtg"])
        row["DRtg"] = float(row["DRtg"])
        row["Team"] = str(year) + " " + row["Team"]
        total_ortg += row["ORtg"]
        total_drtg += row["DRtg"]
        total_teams += 1
        
    average_ortg = total_ortg / total_teams
    average_drtg = total_drtg / total_teams
#     print(average_ortg)
#     print(average_drtg)
#     print(total_teams)
    
    for row in data:
        new_row = {}
        new_row["Team"] = row["Team"]
        new_row["ORtg+"] = round(row["ORtg"] / average_ortg * 100, 1)
        new_row["DRtg+"] = round(average_drtg / row["DRtg"] * 100, 1)
        
        
        if row['\ufeffRk'] == "1":
            new_row["Champions"] = True
            data_champions_CLEANED.append(new_row)
        else:
            new_row["Champions"] = False
        data_playoffs_CLEANED.append(new_row)

print(len(data_playoffs_CLEANED))
display(data_playoffs_CLEANED)
# print(len(data_champions_CLEANED))
# display(data_champions_CLEANED)
```

```python
csv_columns = ["Team", "ORtg+", "DRtg+", "Champions"]

csv_file = "data_playoffs_CLEANED.csv"
try:
    with open(csv_file, 'w') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
        writer.writeheader()
        for row in data_playoffs_CLEANED:
            writer.writerow(row)
except IOError:
    print("I/O error")
```
